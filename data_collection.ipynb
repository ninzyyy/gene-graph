```python
dir = os.path.join(os.path.expanduser('~'), 'Downloads', 'HumanNet-GSP.tsv')
dir
```

```python
data = pd.read_csv(dir, delimiter='\t', header=None)
data.rename(columns={0:'gene_a', 1:'gene_b'}, inplace=True)
data.tail()
```

```python
combined_genes = pd.concat([data['gene_a'], data['gene_b']]).unique()
filtered_df = df[df['Entrez Gene Interactor A'].isin(combined_genes)]
filtered_df = filtered_df.replace('-', None)
filtered_df
```

```python
filtered_df.info()
```

```python
filtered_df['Ontology Term Names'].unique()
```

```python
def standardize_string(s):
    if s is not None:
        parts = sorted(s.split('|'))
        return ','.join(parts)
```

```python
filtered_df = filtered_df[filtered_df['Ontology Term Names'].notna()]
filtered_df['Ontology Term Names'] = filtered_df['Ontology Term Names'].apply(standardize_string)
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```

```python
filtered_df = filtered_df[['Official Symbol Interactor A', 'Official Symbol Interactor B', 'Ontology Term Names']]
filtered_df
```
